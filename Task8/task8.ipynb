```python


import networkx as nx
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

### 8.1 Prepare a code which draw for given network plot of the relation 

```python


G = nx.generators.barabasi_albert_graph(500, 50)
def plot_C(G):
    def avg_degree(G: nx.Graph, k: int) -> float:
        degrees =  dict(G.degree)
        k_nodes = [n for n, v in degrees.items() if v==k] 
        averages = list(sum(degrees[i] for i in G.neighbors(node))/k for node in k_nodes)
        return np.mean(averages)
    X = np.unique(list(dict(G.degree).values()))
    Y = list(map(lambda x: avg_degree(G, x), X))
    plt.plot(X, Y)
    plt.ylim(0, max(Y)*1.2)
plot_C(G)

```
