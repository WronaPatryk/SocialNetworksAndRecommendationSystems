### P1.8 Draw square lattice with your own function. 

```python

class SquareLatticeGraph(nx.Graph):
    # n - number of rows
    # m - number od columns
    def __init__(self, n = 3, m = 4):
        
        super().__init__()
        
        assert n > 0 and m > 0, "n and m should be both positive"
        
        
        # adding nodes
        for i in range(n*m):
            self.add_node(i)
            
        # adding vertical edges
        curr_vertex = 0
        for i in range(m):
            for j in range(n-1):
                self.add_edge(curr_vertex + j, curr_vertex + j + 1)
            curr_vertex += n
            
        # adding horizontal edges
        curr_vertex = 0
        for i in range(n):
            for j in range(0, n*m - n, n):
                self.add_edge(curr_vertex + j, curr_vertex + j + n)
            curr_vertex += 1
        
 
    # Drawing Method
    def drawThisGraph(self):
        print('drawing square lattice graph')
        nx.draw(self)
        plt.show()

```

```python
sth = SquareLatticeGraph(4,4)
```

```python
sth.drawThisGraph()
```

### P1.9 Draw chain with your own function. 

```python

class ChainGraph(nx.Graph):
    # n - number of rows
    def __init__(self, n = 3):
        
        super().__init__()
        
        assert n > 0, "n should be positive"
        
        
        # adding nodes
        for i in range(n):
            self.add_node(i)
            
            
        # adding edges

        for i in range(n-1):
            self.add_edge(i, i + 1)
        
    # Drawing Method
    def drawThisGraph(self):
        print('drawing chain graph')
        nx.draw(self)
        plt.show()

```

```python
G = ChainGraph(7)
G.drawThisGraph()
```

### P1.10 Draw complete graph with your own function. 

```python

class CompleteGraph(nx.Graph):
    # n - number of nodes
    def __init__(self, n = 3):
        
        super().__init__()
        
        assert n > 0, "number of nodes should be positive"
        
        # adding nodes

        for i in range(n):
            self.add_node(i)
        
        # adding edges
        for i in range(n-1):
            for j in range(i + 1, n):
                self.add_edge(i,j)
            
            

        
    # Drawing Method
    def drawThisGraph(self):
        print('drawing complete graph')
        nx.draw(self)
        plt.show()

```

```python
G = CompleteGraph(7)
G.drawThisGraph()
```
